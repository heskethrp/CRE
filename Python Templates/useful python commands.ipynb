In Process Fluid Transport we have an ODE with a variable tau_r=tau*r.
Since we only want tau then we take the tau_r and divide by r, but what happens if r=0?
The program would stop running! So we use an if statement to avoid this error and keep the program running.

```python
tau_r=10
r=2
if (r > 0) :
    tau = tau_r / r
else:
    tau = 0
print (tau)
```

Another fundamental tool in programming is to perform repeated actions or loops.  
This is done using <font color="green"> for </font> or <font color="green"> while </font>statements.
Below is a <font color="green"> for </font> loop that repeats the command starting a i=1 and ending at i=4. 
So the <font color="green"> for </font> does i= 1,2,3  when it gets to 4 it goes out of the indent and to the print statement.
https://pythonnumericalmethods.berkeley.edu/notebooks/chapter05.01-For-Loops.html

```python
n = 0
for i in range(1, 4):
    n = n + i
    
print(n)
```

WHAT IS HAPPENING?<br>

First, the function range(1, 4) is generating a list of numbers beginning at 1 and ending at 3. Check the description of the function range and get familiar with how to use it. In a very simple form, it is range(start, stop, step), and the step is optional with 1 as the default.<br>

The variable n is assigned the value 0.<br>

The variable i is assigned the value 1.<br>

The variable n is assigned the value n + i (0+1=1).<br>

The variable i is assigned the value 2.<br>

The variable n is assigned the value n + i (1+2=3).<br>

The variable i is assigned the value 3.<br>

The variable n is assigned the value n + i (3+3=6).<br>

With no more values to assign in the list, the for-loop is terminated with n = 6.<br>



To explore this <font color="green"> for </font> loop in detail put the print(n) in row 4.  Remember to indent.  Do you have line numbers? View Toggle line numbers.

```python
n = 0
for i in range(1, 4):
    n = n + i
    print(i,n)
print(n)
```

The loop below is a simplified version of the loop used in the secant method to find the zero of a function.  I have added a safety switch in here that prevents the loop from running too many times.  I use an <font color="green"> and </font> in the logic statement to have the loop run until either the err is less than 1e-6 or the loop runs 20 times.  And an <font color="green"> if  </font> statement is at the end letting the user know what happened!

```python
err=1e-5
k=0
while (err>1e-6 and k<20): #multiple iterations to get err<1e-6
    k=k+1
    err = err/k
    print(k,err)

if err<=1e-6:
    print('The err is <= 1e-6 err=', err)
else:
    print('too many iterations k=',k) 
```

```python
print('{:f}{:8.2f}{:8.2f}{:10.2E}'.format(x2, x2, x3,x4)) #This is the new style format.  The 8 will fill the number, including the decimal point into 8 spaces with 8 numbers to the right of the decimal point.
```
